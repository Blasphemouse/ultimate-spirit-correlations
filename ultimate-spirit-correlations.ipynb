```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
menResults2018=pd.read_csv('2018-Results-Men.csv', sep=',',header=0)
menSpirit2018=pd.read_csv('2018-Spirit-Men.csv', sep=',',header=0)
menSeed2018=pd.read_csv('2018-Seed-Men.csv', sep=',',header=0)
menPoints2018=pd.read_csv('2018-Points-Men.csv', sep=',',header=0)

mixedResults2018=pd.read_csv('2018-Results-Mixed.csv', sep=',',header=0)
mixedSpirit2018=pd.read_csv('2018-Spirit-Mixed.csv', sep=',',header=0)
mixedSeed2018=pd.read_csv('2018-Seed-Mixed.csv', sep=',',header=0)
mixedPoints2018=pd.read_csv('2018-Points-Mixed.csv', sep=',',header=0)

womenResults2018=pd.read_csv('2018-Results-Women.csv', sep=',',header=0)
womenSpirit2018=pd.read_csv('2018-Spirit-Women.csv', sep=',',header=0)
womenSeed2018=pd.read_csv('2018-Seed-Women.csv', sep=',',header=0)
womenPoints2018=pd.read_csv('2018-Points-Women.csv', sep=',',header=0)
```

```python
men2018 = pd.merge(menResults2018, menSpirit2018, left_on = 'Team', right_on = 'Team')
men2018 = pd.merge(men2018, menSeed2018, left_on = 'Team', right_on = 'Team')
men2018['SeedMinusRank'] = men2018['Seed'] - men2018['Rank']
men2018 = pd.merge(men2018, menPoints2018, left_on = 'Team', right_on = 'Team')

mixed2018 = pd.merge(mixedResults2018, mixedSpirit2018, left_on = 'Team', right_on = 'Team')
mixed2018 = pd.merge(mixed2018, mixedSeed2018, left_on = 'Team', right_on = 'Team')
mixed2018['SeedMinusRank'] = mixed2018['Seed'] - mixed2018['Rank']
mixed2018 = pd.merge(mixed2018, mixedPoints2018, left_on = 'Team', right_on = 'Team')

women2018 = pd.merge(womenResults2018, womenSpirit2018, left_on = 'Team', right_on = 'Team')
women2018 = pd.merge(women2018, womenSeed2018, left_on = 'Team', right_on = 'Team')
women2018['SeedMinusRank'] = women2018['Seed'] - women2018['Rank']
women2018 = pd.merge(women2018, womenPoints2018, left_on = 'Team', right_on = 'Team')
```

```python
all2018 = pd.concat([men2018, mixed2018, women2018])
```

* In Men's and Mixed, a high rank (i.e. lower placement) indicates a moderate positive correlation
    * In Men's, there is an exception in the Rules category which has a very slight negative correlation.
    * In Women's, the relationship between rank is significantly less pronounced. (Thanks Fury & Molly Brown!)
* In Women's, there is a moderate negative correlation between communication and fouls.
    * Wildfire and Schwa both had (slightly) below average on Fouls, but above average on Communication.
* Overall, Fouls+Fair+Attitude have the largest impact on a larger Spirit score, but Communication and then Rules are not far behind.
* Seed and Spirit don't have too much of a relationship; however, Seed and Rules have an inverse relationship (i.e. top teams were, at least, perceived to have a solid grasp of the rules and stayed on time). Certainly for timekeeping, top teams would benefit from having a coach. (In Men's, all teams (except High Five and Sub Zero) have a coach listed on the team page. Potentially this would have an effect on spirit alone - but for timekeeping, they would need to be present -- no easy way to know who attended.)
* SeedMinusRank (positive = performed better than expected) does have a moderate inverse relationship across the board, strongest in Mixed; also strong in Women's Rules+Fouls but a moderate positive relationship in Women's Communication.
* Point Differential Sum and Average are almost identical to each other; however the sum of the absolute values does seem a bit more clear. This data point is the valuable one of the three since it measures what we actually want. Both the winner and loser of lopsided games will have a higher "PointDiffAbsSum" - and as we suspected, they receive higher spirit scores as a result.
